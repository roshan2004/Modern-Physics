### Photoelectric effect
In 1887, Heinrich Hertz discovered a phenomenon called photoelectric effect in which by shining light having suitable frequency on the metallic surface causes electrons called photoelectons to be emitted from the surface. Two observations made during this experiment were in contrast with classical theory. First, that the energy of the photoelectrons is independent of the intensity of the incident light and that there is a threshold frquency of incident light below which no electrons are ejected from the surface. These results could not be explained by classical mechanics. Albert Einstein attempted to explain this phenomenon by first quantizing the energy levels of the electrons similar to what Planck did:

$ E = nhf \tag{1} $

Einstein took this one step further and suggested that the light energy emitted was also quantized packets of energy called photons. It leads to the conclusion that the kinetic energy of emitted light will have the following form:

$KE = hf - hf_0. \tag{2}$

This result was able to successfully explain the experiment. The above equation demonstrates that the kinetic energy of the emitted electron will be equal to the energy (not intensity) of the incident light minus a threshold energy which is also called a work function.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def define_figure(xlabel="X",ylabel="Y"):
    # setup plot parameters
    fig = plt.figure(figsize=(8,6), dpi= 300, facecolor='w', edgecolor='k')
    ax = plt.subplot(111)
    ax.grid(b=True, which='major', axis='both', color='#808080', linestyle='--')
    ax.set_xlabel(xlabel,size=15)
    ax.set_ylabel(ylabel,size=15)
    plt.tick_params(axis='both',labelsize=15)
    return ax
# Photoelectric effect for Lithium
h = 6.626e-34  # converted unites
c = 3.0E8      # m/s

f0 = 3.75e14 # Threshold frequency for Lithium

#Einstein's Photoelectric equation
def photoelectric(x):
    return h*x - h*f0  #h*f0 = Work function

# initialize a figure
ax = define_figure(xlabel='$ f (10^{15} Hz)$',ylabel="KE$/10^{-19}$ J")
# make an array containing domain of wavelengths to consider
x = np.linspace(0.0,15e14,1000)

# plot quantum result
ax.plot(x,photoelectric(x),lw=2)
# make legend
#ax.legend(fontsize=10,markerscale=0.5)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.ticklabel_format(useOffset=False, style = 'sci')
plt.savefig('photoelectric.png')

```
