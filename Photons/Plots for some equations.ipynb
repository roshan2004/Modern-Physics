```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def define_figure(xlabel="X",ylabel="Y"):
    # setup plot parameters
    fig = plt.figure(figsize=(10,8), dpi= 300, facecolor='w', edgecolor='k')
    ax = plt.subplot(111)
    ax.grid(b=True, which='major', axis='both', color='#808080', linestyle='--')
    ax.set_xlabel(xlabel,size=20)
    ax.set_ylabel(ylabel,size=20)
    plt.tick_params(axis='both',labelsize=20)
    return ax
# Photoelectric effect for Lithium
h = 6.626e-34  # converted unites
c = 3.0E8      # m/s

v0 = 3.75e14 # Threshold frequency for Lithium

#Einstein's Photoelectric equation
def photoelectric(x):
    return h*x - h*v0  #h*v0 = Work function

# initialize a figure
ax = define_figure(xlabel='$ v (10^{15}s^{-1})$',ylabel="KE$/10^{-19}$ J")
# make an array containing domain of wavelengths to consider
x = np.linspace(0.0,15e14,1000)

# plot quantum result
ax.plot(x,photoelectric(x),lw=3)
# make legend
#ax.legend(fontsize=10,markerscale=0.5)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.ticklabel_format(useOffset=False, style = 'sci')
plt.savefig('photoelectric.png')

```
